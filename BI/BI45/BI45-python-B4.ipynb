```python
import pandas as pd
from tqdm import tqdm
from datetime import datetime
import warnings
import numpy as np
warnings.filterwarnings('ignore') # turn off warnings
```

```python
def show_values(x):
    return list(set(x))
```

```python
df = pd.read_csv(r"C:\Users\hieudd\Downloads\netflix_titles.csv")

df['director'] = df['director'].fillna(value='Unknow')
df['cast'] = df['cast'].fillna(value='Unknow')

df['date_added'] = df['date_added'].str.strip()
df['date_added'] = df.apply(lambda x: 'January 1, ' + str(x['release_year']) if pd.isna(x['date_added']) == True else x['date_added'], axis=1)
df['date_added'] = df['date_added'].apply(lambda x: datetime.strptime(x, '%B %d, %Y').strftime('%Y-%m-%d'))

df['duration'] = df.apply(lambda x: x['rating'] if pd.isna(x['duration']) == True else x['duration'], axis=1)
df['rating'] = df.apply(lambda x : 'Unknow' if 'min' in str(x['rating']) else x['rating'], axis=1)
df['rating'].fillna(value='Unknow',inplace=True)
df
```

```python
for i in tqdm(show_values(df['director'])):
    try:
        df.loc[df['director'] == i, 'country'] = df.loc[df['director'] == i, 'country'].fillna(method='bfill')
        
    except:
        df.loc[df['director'] == i, 'country'] = df.loc[df['director'] == i, 'country'].fillna(method='ffill')

df['country'].fillna(value='Unknow',inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

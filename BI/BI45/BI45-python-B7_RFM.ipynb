```python
import pandas as pd 
from datetime import datetime
import os


pathfolder = r'C:\Users\hieudd\Downloads\folder'
```

```python
pro = pd.read_csv(os.path.join(pathfolder,'AdventureWorks_Products.csv'))
pro
```

```python
sale15 = pd.read_csv(os.path.join(pathfolder,'AdventureWorks_Sales_2015.csv'))
sale16 = pd.read_csv(os.path.join(pathfolder,'AdventureWorks_Sales_2016.csv'))
sale17 = pd.read_csv(os.path.join(pathfolder,'AdventureWorks_Sales_2017.csv'))
soh = pd.concat([sale15,sale16])
soh = pd.concat([soh,sale17])
soh = soh.merge(pro[['ProductKey','ProductPrice']],how='inner',on='ProductKey')
soh['OrderDate'] = pd.to_datetime(soh['OrderDate'])
soh['StockDate'] = pd.to_datetime(soh['StockDate'])
soh['Today'] = pd.to_datetime(datetime.now().strftime('%Y-%m-%d'))
soh['Datediff'] = soh['Today'] - soh['StockDate']
soh['Datediff'] = soh['Datediff'].astype('str').str.replace(' days','')
soh['Datediff'] = soh['Datediff'].astype('int')
soh['SaleAmount'] = soh['OrderQuantity'] * soh['ProductPrice']
soh
```

```python
soh.info()
```

```python
RFM = soh.groupby(['CustomerKey']).agg(R=('Datediff','max'),
                                       F=('OrderNumber','nunique'),
                                       M=('SaleAmount','sum')).reset_index()

Fsorce = RFM[['F']].drop_duplicates()
Fsorce['F_socre'] = pd.qcut(Fsorce['F'].drop_duplicates(),q=4,labels=[1,2,3,4])
Fsorce
```

```python

RFM['R_socre'] = pd.qcut(RFM['R'], q=4, labels= [1,2,3,4])
RFM['M_socre'] = pd.qcut(RFM['M'], q=4, labels= [1,2,3,4])
RFM = RFM.merge(Fsorce, on='F')
RFM['RFM_socre'] = RFM['R_socre'].astype('str') + RFM['M_socre'].astype('str') + RFM['M_socre'].astype('str')
RFM
```

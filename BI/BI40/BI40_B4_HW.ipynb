# Xử lý những cột còn lại:

- Dữ liệu còn lại có thể dựa vào cột Ngày tiếp nhận hồ sơ chia làm 3 loại dữ liệu:

    + Loại 1: dữ liệu đúng cấu trúc và thiếu dữ liệu

    + Loại 2: dự liệu trong cột điền sai giá trị 

    + Loại 3: dữ liệu thiếu giá trị 

- Xử lý từng loại dữ liệu và tiến hành ghép dữ liệu để tạo ra dữ liệu cuối cùng.

## Loại 1: dữ liệu đúng cấu trúc và thiếu dữ liệu

- Prework: cột này đã được nói tới trong data dictionary do vậy không xử lý

- Họ và tên, Nguồn: thay thế bằng giá trị 'Unknow' 

- Đặc điểm ứng viên: cột này được sử dụng sai mục đích khá nhiều, tuy nhiên lượng dữ liệu trong đó vẫn có do vậy giữ nguyên giá trị, thay thế giá trị null = 'Unknow'

- Ví trí có thể phù hợp	và Tình trạng: 

    + Nếu Tình trạng = PassCV => fillna Ví trí có thể phù hợp = Processing

    + Nếu Ví trí có thể phù hợp không null => Tình trạng fillna = PassCV

    + Nếu cả 2 cùng null => 'Watting'

- Loại 2, Loại 3 sẽ làm tương tự trừ những trường hợp dữ liệu cần xử lý đặc biệt

```python
L1 = df[(df['Ngày tiếp nhận hồ sơ'].str.contains('00:00:00')==True)]
L1['Họ và tên'].fillna(value='Unknow',inplace=True)
L1['Nguồn'].fillna(value='Unknow',inplace=True)
L1['Đặc điểm ứng viên'].fillna(value='Unknow',inplace=True)
L1
```

```python
L1['Ví trí có thể phù hợp'] = L1.apply(lambda x: 'Processing' if (x['Tình trạng']=='Pass CV') else x['Ví trí có thể phù hợp'], axis=1)
L1['Tình trạng'] = L1.apply(lambda x: 'PassCV' if (pd.isna(x['Ví trí có thể phù hợp'])==False) else x['Tình trạng'], axis=1)
```

```python
L1['Ví trí có thể phù hợp'].fillna(value='Watting',inplace=True)
L1['Tình trạng'].fillna(value='Watting',inplace=True)
```

```python
L1
```

```python
L1.info()
```

## Loại 2: Dữ liệu cột điền sai giá trị

```python
L2 = df[df['Ngày tiếp nhận hồ sơ'].str.contains('00:00:00')==False]
L2['Đặc điểm ứng viên'].fillna(value='Unknow',inplace=True)
L2['Nguồn'].fillna(value='Unknow',inplace=True)
L2
```

```python
L2['Họ và tên'] = L2['Ngày tiếp nhận hồ sơ'].where(pd.isna(L2['Họ và tên'])==True,L2['Họ và tên'])
L2['Tình trạng'] = L2.apply(lambda x: 'PassCV' if (pd.isna(x['Ví trí có thể phù hợp'])==False) else x['Tình trạng'], axis=1)
L2['Ví trí có thể phù hợp'] = L2.apply(lambda x: 'Processing' if (x['Tình trạng']=='Pass CV') else x['Ví trí có thể phù hợp'], axis=1)
L2['Ví trí có thể phù hợp'] = L2.apply(lambda x : 'FALSE' if (x['Tình trạng']=='FALSE') else x['Ví trí có thể phù hợp'],axis=1)
L2['Ví trí có thể phù hợp'].fillna(value='Watting',inplace=True)
L2['Tình trạng'].fillna(value='Watting',inplace=True)
L2['Ngày tiếp nhận hồ sơ'] = 'Unknow'
L2
```

## Loại 3: 

```python
L3 = df[(df['Ngày tiếp nhận hồ sơ'].isnull())]
L3['Ngày tiếp nhận hồ sơ'].fillna(value='Unknow',inplace=True)
L3['Họ và tên'].fillna(value='Unknow',inplace=True)
L3['Đặc điểm ứng viên'].fillna(value='Unknow',inplace=True)
L3['Nguồn'].fillna(value='Unknow',inplace=True)
L3
```

```python
L3['Ví trí có thể phù hợp'] = L3.apply(lambda x: 'Processing' if (x['Tình trạng']=='Pass CV') else x['Ví trí có thể phù hợp'], axis=1)
L3['Tình trạng'] = L3.apply(lambda x: 'PassCV' if (pd.isna(x['Ví trí có thể phù hợp'])==False) else x['Tình trạng'], axis=1)
L3['Ví trí có thể phù hợp'].fillna(value='Watting',inplace=True)
L3['Tình trạng'].fillna(value='Watting',inplace=True)
```

```python
df = pd.concat([L1,L2,L3])
df
```

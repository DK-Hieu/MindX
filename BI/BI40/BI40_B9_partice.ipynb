```python
import pandas as pd
```

```python
def values_show(x):
    return list(set(x))
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTgQfVdqjCcmQpx9KTBxv-47cbuzq035YMq47Awd889gpIMAlpBlxGqH4ZTBXtPvg/pub?output=csv')
df['InvoiceDate'] = df['InvoiceDate'].str.split(' ', expand=True)[0]
df['month'] = df['InvoiceDate'].str.split('/', expand=True)[0]
df['day'] = df['InvoiceDate'].str.split('/', expand=True)[1]
df['year'] = df['InvoiceDate'].str.split('/', expand=True)[2].replace('10','2010').replace('11','2011')
df['date'] = df['year'] + '-' + df['month'] +'-'+ df['day']
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df
```

```python
df1 = df[['date','CustomerID']].drop_duplicates()
df1.reset_index(drop=True,inplace=True)
# df1.set_index('date',inplace=True)
df1
```

## Kiểm tra và xử lý dữ liệu null 

- Ta tiến hành xử lý dữ liệu từ cột có số lượng giá trị null ít nhất

### duration

- 3 giá trị null có giá trị bị lệch cột

- giá trị của cột đan nằm trong cột rating

- ta sẽ thay thế null = giá trị trong cột rating

- giá trị tại cột rating sẽ thay thế bằng giá trị orther

```python
df[df['duration'].isnull()]
```

```python
df['duration'] = df['rating'].where(df['duration'].isnull()==True,df['duration'])
df[df['show_id'].isin(['s5542','s5795','s5814'])]
```

```python
df['rating'] = df['rating'].apply(lambda x: 'Other' if 'min' in str(x) else x)
df[df['show_id'].isin(['s5542','s5795','s5814'])]
```

```python
df.info()
```

### director, cast

- Dữ liệu này không thể tìm kiếm các cột khác tương ứng lại chứa dữ liệu => thay thế bằng giá trị unknown


```python
df['director'].fillna(value='unknow',inplace=True)
df['cast'].fillna(value='unknow',inplace=True)
df.info()
```

### country

- Dữ liệu country có thể làm theo kiểu đơn gian là ta cũng thay dữ liệu null bằng giá trị unknow do lượng dữ liệu thiếu khá lớn

- Tuy nhiên trong trường hợp cần cứu dữ liệu tối đa có thể ta có thể chia dữ liệu null làm 2 hoặc 3 loại dữ liệu ở đây mình sẽ chia làm 2 loại:

    + Loại 1: dữ liêu null ở cả director và country => với dữ liệu này ta chỉ có thể thay bằng unknow

    + Loại 2: dữ liệu chỉ null ở country và có dữ liệu director, tại đây ta sẽ thay dữ liệu null bằng dữ liệu country của một phim khác có cùng director
    
      độ chính xác tầm 80%


### Loại 1:

```python
df[(df['country'].isnull())&(df['director']=='unknow')]
```

```python
df['country'] = df['director'].where(df['director']=='unknow',df['country'])
```

```python
df[df['country']=='unknow']
```

### Loại 2:

```python
df[(df['country'].isnull())]
```

```python
# lấy ra list danh sách director cần sửa
directorlist = list(set(df[(df['country'].isnull())]['director']))
```

```python
df[(df['director']=='Julien Leclercq') & (df['country'].isnull())==False]['country'].iloc[0]
```

```python
for i in directorlist:
    a = df[(df['director']==i) & (df['country'].isnull())==False]['country'].iloc[0] # Lấy ra giá trị country tương ứng với director
    df['country'] = df.apply(lambda x: a if ((x['director'] ==i) & 
                                             (pd.isna(x['country'])==True)) else x['country'], axis=1)
```

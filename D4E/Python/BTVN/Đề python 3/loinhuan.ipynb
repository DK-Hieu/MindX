```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
import requests
import pickle
print('Completed import lib')
```

```python
loinhuan = pd.read_csv(r'D:\FPTS\FPTS\loinhuan.csv')
loinhuan.drop(columns='Unnamed: 0', inplace=True)
loinhuan
```

```python
list21 = list(loinhuan[loinhuan['year']==2021]['ticker'])
df_lnctm = loinhuan[loinhuan['year']==2021].groupby('Industry3')[['pl_lnctm']].sum().reset_index()
df_lnctm.rename(columns={'pl_lnctm':'pl_lnctm_2021'},inplace=True)
df_lnctm
```

```python
def calcultion_a(year):
    lnctm_a = loinhuan[loinhuan['year']==year].groupby('Industry3')[['pl_lnctm']].sum().reset_index()
    lnctm_a.rename(columns={'pl_lnctm':'pl_lnctm_a_%s'%year},inplace=True)
    return lnctm_a

def calcultion_a1(year):
    df = loinhuan[loinhuan['year']==2021]
    lista = list(loinhuan[loinhuan['year']==year]['ticker'])
    lnctm_a1 = df[(df['ticker'].isin(list21))&(df['ticker'].isin(lista))].groupby('Industry3')[['pl_lnctm']].sum().reset_index()
    lnctm_a1.rename(columns={'pl_lnctm':'pl_lnctm_a1_%s'%year},inplace=True)
    return lnctm_a1
    
```

```python
for i in range(2008,2022):
    a = calcultion_a(i)
    a1 = calcultion_a1(i)
    df_lnctm = df_lnctm.merge(a, on='Industry3',how='left')
    df_lnctm = df_lnctm.merge(a1, on='Industry3',how='left')
    df_lnctm['x_%s' %i] = (df_lnctm['pl_lnctm_a_%s' %i] *df_lnctm['pl_lnctm_2021'])/df_lnctm['pl_lnctm_a1_%s' %i]
df_lnctm
```

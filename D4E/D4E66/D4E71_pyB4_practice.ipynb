```python
df1 = pd.read_csv(r"D:\MindX\D4E\mexico-real-estate-1.csv")
df2 = pd.read_csv(r"D:\MindX\D4E\mexico-real-estate-2.csv")
df3 = pd.read_csv(r"D:\MindX\D4E\mexico-real-estate-3.csv")
```

### df1:

```python
df2.rename(columns={'price_mxn':'price_usd'},inplace=True)
df2['price_usd'] = df2['price_usd'] / 17.35
df2
```

### df3:
- place_with_parent_names: tách dữ liệu và lấy cột state => split
- lat-lon: tách dữ liệu làm 2 cột lat lon => split

```python
df3
```

```python
df3['state'] = df3['place_with_parent_names'].str.split('|',expand=True)[3]
df3['lat'] = df3['lat-lon'].str.split(',',expand=True)[0].astype('float')
df3['lon'] = df3['lat-lon'].str.split(',',expand=True)[1].astype('float')
df31 = df3[['property_type', 'state','lat','lon','area_m2','price_usd']]
df31
```

```python
df31.info()
```

```python
c1 = df1.append(df2)
c1 = c1.append(df31)
c1
```

```python
c2 = pd.concat([df1,df2])
c2 = pd.concat([c2,df31])
c2
```

### Câu 1: Bạn hãy xây dựng một hàm nhận tham số truyền vào là 1 chuỗi ký tự bất kỳ. Hàm kiểm tra và trả về kết quả chuỗi trên là một đoạn password hợp lệ hay không, biết rằng password hợp lệ cần thoả mãn các tiêu chí: 
- Chứa 1 ký tự đặc biệt trong các ký tự !, @, #, $, ^ 
- Chứa ít nhất một ký tự số
- Chứa ít nhất một ký tự chữ cái viết hoa
- Chứa ít nhất một ký tự chữ cái viết thường
- Độ dài từ 8 - 15 ký tự. (0.75đ)


```python
def dodai(x):
    if len(x) > 8 and len(x) < 15:
        return True
    else:
        return False
    
def kytuDB(x):
    special_characters = "!@#$^"
    for char in x:
        if char in special_characters:
            return True
    return False

def kytuNumber(x):
    special_characters = "0123456789"
    for char in x:
        if char in special_characters:
            return True
    return False

def kytuUpper(x):
    return any(char.isupper() for char in x)

def kytuLowwer(x):
    return any(char.islower() for char in x)

password = input()

if dodai(password) == True and kytuDB(password) == True and kytuNumber(password) == True and kytuUpper(password) == True and kytuLowwer(password):
    print('Password hợp lệ')
    print(password)
else:
    print("Password không hợp lệ, cần độ dài 8 đến 15 ký tự, có ít nhất 1 số, 1 chữ hoa, 1 chữ thường, 1 ký tự đặc biệt gồm (!,@,#,$,^)")
    print(password)
```

```python
import pandas as pd
```

```python
""" Bạn hãy lấy ra các đơn hàng thuộc Branch A hoặc C (0.25đ) """
df[(df['Branch']=='A')|(df['Branch']=='C')]
```

```python
""" Cột InvoiceID, giá trị được lưu trữ theo dạng XXX-StoreID-XXXX. Bạn hãy tạo 1 dataframe mới gồm các cột như trên file và tạo thêm một cột StoreID được tách từ InvoiceID. Sau đó tính tổng doanh số của từng StoreID đó. (0.25đ) """
df1 = df.copy()
df1['StoreID'] = df1['Invoice ID'].str.split('-',expand=True)[1]
df1['Revene'] = df1['Unit price'] * df1['Quantity']
df1 = df1.groupby('StoreID').agg(total_sale = ('Revene','sum')).reset_index()
df1
```

### Câu 3:  Bạn hãy xây dựng một hàm nhận tham số truyền vào là một danh sách bất kỳ chỉ chứa các số. Hàm tính toán và  trả về số lớn nhì trong danh sách. Với trường hợp danh sách gồm các số bằng nhau, hàm trả về số lớn nhất. Nếu có nhiều giá trị bằng số lớn nhì trong danh sách, hàm trả về 1 giá trị duy nhất. (1.5đ)
- Ví dụ: - Với danh sách [1,1,1,1,1,1] -> Hàm trả về 1
- Với danh sách [1,2,1,1,2,3] -> Hàm trả về 2

```python
ds = input('Nhập dãy số: ')

list_ds = [eval(i) for i in ds]
list_ds1 = list(set(list_ds))
list_ds1.sort(reverse=True)
print('dãy số nhập vào: ', ds)
if len(list_ds1) == 1: 
    print('Số lớn mhất: ',list_ds1[0])
else:
    print('Số lớn thứ 2: ',list_ds1[1])
```
